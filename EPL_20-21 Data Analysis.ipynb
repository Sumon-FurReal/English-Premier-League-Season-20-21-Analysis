```python
#Imports 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loding Dataset 

epl_df = pd.read_csv('C:\\Users\\Dell\\Documents\\Python Scripts\\EPL_20_21.csv')
epl_df.head()
```

```python
# Dataset info

epl_df.info()
```

```python
# Summary

epl_df.describe()
```

```python
# checking null 

epl_df.isna().sum()
```

```python
# New col - Min/match and goal/match

epl_df['Mins_Per_Match'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['Goals_Per_Match'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)
epl_df.head()
```

```python
# Total Goals Scored in 2020-2021 season

Total_Goals = epl_df['Goals'].sum()
Total_Goals
```

```python
# Penalty Goals 

Total_PenaltyGoals = epl_df['Penalty_Goals'].sum()
Total_PenaltyGoals
```

```python
# Penalty Attempts

Total_PenaltyAttempts = epl_df['Penalty_Attempted'].sum()
Total_PenaltyAttempts
```

```python
# Pie Chart for penalties socred vs missed 

plt.figure(figsize=(13,6))
Penalties_Not_Scored = epl_df['Penalty_Attempted'].sum() - Total_PenaltyGoals
data =  [Penalties_Not_Scored, Total_PenaltyGoals]
lab = ['Penalties Missed', 'Penalties Scored']
col = sns.color_palette('hls')
plt.pie(data, labels = lab, colors = col, autopct = '%.0f%%')
plt.show()
```

```python
# Unique Positions 

epl_df['Position'].unique()
```

```python
# Total Forward Players 

epl_df[epl_df['Position'] == 'FW']
```

```python
# EPL plaers from different countries 

np.size((epl_df['Nationality'].unique()))
```

```python
# Most players from which countries ?

nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize=(12,6), color = sns.color_palette("hls"))
```

```python
# Clubs with Max Players in Squad

epl_df['Club'].value_counts().nlargest(10).plot(kind = 'bar', color = sns.color_palette('hls'))
```

```python
# Clubs with Least Players in Squad

epl_df['Club'].value_counts().nsmallest(10).plot(kind = 'bar', color = sns.color_palette('hls'))
plt.title("Clubs with Least Players in Squad", fontsize = 20)
```

```python
# Players Based on age group

players_ageing_under20 = epl_df[epl_df['Age'] <= 20]
players_ageing_betn20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
players_ageing_betn25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
players_ageing_above30 = epl_df[epl_df['Age'] >= 30]
```

```python
# Pie Chart based on age group 

p = np.array([players_ageing_under20['Name'].count(),players_ageing_betn20_25['Name'].count(),players_ageing_betn25_30['Name'].count(),players_ageing_above30['Name'].count()])
lab = ["<=20",">20 & <=25", ">25 & <=30", ">30"]
plt.title('Total Players with Age Groups', fontsize = 20)
plt.pie(p, labels = lab, autopct = "%.0f%%")
plt.show()
```

```python
# Total Under 20 players in each club 

players_under20 = epl_df[epl_df['Age'] <20 ]
players_under20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette('magma'))
plt.title('Total Under 20 players in each clubs', fontsize = 15)
```

```python
# Under 20 Plaers in ManU

players_under20[players_under20["Club"] == "Manchester United"]
```

```python
# Under 20 Plaers in ManU

players_under20[players_under20["Club"] == "Chelsea"]
```

```python
# Average age of players in each club 

plt.figure(figsize=(12,6))
sns.boxplot(x = 'Club', y='Age', data = epl_df)
plt.xticks(rotation = 90 )
plt.title("Average age of players in each club" , fontsize = 15)
```

```python
# Average age of players in each club (verificaiton)

num_player = epl_df.groupby("Club").size()
dat = (epl_df.groupby("Club")["Age"].sum()) / num_player
dat.sort_values(ascending = False)
```

```python
# Total assist from each club 

assists_by_clubs = epl_df.groupby('Club', as_index = False)['Assists'].sum()
sns.color_palette("tab10")
ax = sns.barplot(x = 'Club', y = 'Assists', data = assists_by_clubs.sort_values(by='Assists'), palette = "hls")
ax.set_xlabel("Club", fontsize = 15)
ax.set_ylabel("Assists", fontsize = 15)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of clubs vs total assists', fontsize = 15)
```

```python
# Top 10 Players with most assists 

top10_assists = epl_df[["Name","Club","Assists","Matches"]].nlargest(n=10, columns = "Assists")
top10_assists
```

```python
# Total goals from each club 

goals_by_clubs = epl_df.groupby('Club', as_index = False)['Goals'].sum()
sns.color_palette("tab10")
ax = sns.barplot(x = 'Club', y = 'Goals', data = goals_by_clubs.sort_values(by='Goals'), palette = "rocket")
ax.set_xlabel("Club", fontsize = 15)
ax.set_ylabel("Goals", fontsize = 15)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize'] = (8,5)
plt.title('Plot of clubs vs total goals', fontsize = 15)
```

```python
# Most Goals by players

top_10_goals = epl_df[["Name","Club","Goals","Matches"]].nlargest(n=10 , columns = "Goals")
top_10_goals
```

```python
# Goals per match 

top_10_goals_per_match = epl_df[["Name","Club", "Goals_Per_Match","Matches","Goals"]].nlargest(n=10, columns = "Goals_Per_Match")
top_10_goals_per_match
```

```python
# Pie chart - Goals with and without assists 

plt.figure(figsize=(10,5))
assists = epl_df["Assists"].sum()
dat = [Total_Goals - assists, assists]
lab = ["Goals without assists", "Goals with assists"]
col = sns.color_palette("hls")
plt.pie(dat, labels = lab, colors = col, autopct = "%.0f%%")
plt.show()
```

```python
# top 20 players with most yellow cards  

epl_yellow = epl_df.sort_values(by="Yellow_Cards", ascending=False)[:20]
plt.figure(figsize=(10,5))
plt.title("Players with most yellow cards", fontsize = 15)
yc = sns.barplot(x=epl_yellow["Name"], y=epl_yellow["Yellow_Cards"], label="Players", color = "yellow")
plt.ylabel("Number of yellow cards")
yc.set_xticklabels(yc.get_xticklabels(), rotation = 90)
yc
```

```python
# top 20 players with most red cards  

epl_red = epl_df.sort_values(by="Red_Cards", ascending=False)[:20]
plt.figure(figsize=(10,5))
plt.title("Players with most red cards", fontsize = 15)
rc = sns.barplot(x=epl_red["Name"], y=epl_red["Red_Cards"], label="Players", color = "red")
plt.ylabel("Number of red cards")
rc.set_xticklabels(rc.get_xticklabels(), rotation = 90)
rc
```
